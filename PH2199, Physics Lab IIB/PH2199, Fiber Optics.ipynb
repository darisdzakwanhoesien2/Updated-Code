```python

import pandas as pd
import numpy as np
import math
```

```python

lens = [0.1,0.15,0.2,0.3,0.4,0.6,1.3,2,2.5]
output_raw_1 = [1.290,1.119,0.979,0.843,0.569,0.416,0.087,0.020,0.008] 
output_raw_2 = [1.339,1.162,1.01,0.881,0.593,0.424,0.087,0.020,0.008] 
baseline = 0.002
```

```python
input_val = np.array([1.72,1.80])
input_mean = input_val.mean()
input_error = round(input_val[1] - input_mean,3)
input_mean, input_error
```

3.2 Characterization of the Laser and Photometer

```python

data = pd.DataFrame()
data["lens"] = lens
data["output_raw_1"] = output_raw_1
data["output_raw_2"] = output_raw_2
data
```

```python
data['Raw Value'] = (data["output_raw_1"] + data["output_raw_2"])/2 #Average
data['error'] = (data["output_raw_2"] - data["output_raw_1"])/2 #Error
data["Value"] = data['Raw Value'] - baseline #Real Value
data['Ratio'] = input_mean/data["Value"] #Ratio
data["Log Ratio"] = data['Ratio'].apply(np.log10) #Attenuation
data["Error Log Ratio"] = abs(data["Log Ratio"]-lens)*100/lens #Attenuation Error
data
```

```python
data_std = round(data["Error Log Ratio"].std(),3) #Total Attenuation Error 
data_std
```

3.3 Waveguiding properties of an optical fiber

```python

single_line = pd.DataFrame()

length = [35.75,26.60,25.65,23.4,21.3] 
power1 = [66.9,41.9,38.4,37.3,62.9]
power2 = [68.1,43.9,38.4,37.3,62.9]

single_line["Length"] = length
single_line["Power 1"] = power1
single_line["Power 2"] = power2

single_line
```

```python

single_line['Power'] = (single_line["Power 2"] + single_line["Power 1"])/2
single_line['Error'] = (single_line["Power 2"] - single_line["Power 1"])/2
single_line['Value'] = single_line['Power'] - baseline
single_line["Attentuation"] = input_mean/single_line['Value']
single_line["Log Attentuation"] = -single_line["Attentuation"].apply(np.log)
single_line
```
