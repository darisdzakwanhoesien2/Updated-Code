```python
import numpy as np
x = np.linspace(-3,3,1000)
y = np.sqrt(9 - x**2)
z = -y
import matplotlib.pyplot as plt
plt.plot(x,y)
plt.plot(x,z)
plt.figure(figsize=(16,10))
plt.show
```

```python
def squircle(r,s,label):
    # https://en.wikipedia.org/wiki/Squircle
    
    x = np.linspace(-3,3,1000)
    coef = 1 - (s**2)*(x**2)/r**2
    y_ = np.sqrt((1-x**2)/coef)

    plt.plot(x,y_)
    plt.plot(x,-y_)
    plt.xlim(-label,label)
    plt.ylim(-label,label)
    plt.figure(figsize=(16,10))
    plt.show()
for i in range(11):
    print(3*i/10)
    squircle(3,3*i/10,1.6)
```

```python
squircle(3,2.5,1.8)
```

```python
for i in range(10):
    n = 1+i
    r = 3
    x_ = np.linspace(3,3,100)
    y_lane = (r**n - x**n)**(1/n)

    import matplotlib.pyplot as plt
    plt.plot(x,y_lane)
    plt.plot(x,-y_lane)
    plt.title('x^' + str(n) + ' + y^' + str(n) + ' = z^' + str(n) )
    plt.show()
```

```python
for i in range(100):
    n = 2*(1+i)/10
    r = 3
    x_ = np.linspace(3,3,100)
    y_lane = (r**n - x**n)**(1/n)

    import matplotlib.pyplot as plt
    plt.plot(x,y_lane)
    plt.plot(x,-y_lane)
    plt.plot(-x,y_lane)
    plt.plot(-x,-y_lane)
    plt.title('x^' + str(n) + ' + y^' + str(n) + '=z^' + str(n) )
    plt.show()
```
