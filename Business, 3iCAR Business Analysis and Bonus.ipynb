Skema Bonus 3iCAR
Bonus Titik/Persistensi, bonus yang diberikan di tahun kedua, perhitungannya adalah 1000/2000/2800 per downline
Ada beberapa tipe bonus yang dimiliki tabungan 3iCAR, yaitu:

Bonus Rekrut, yang dimana kamu mendapatkan komisi jika mengajak orang berinvestasi di 3iCAR
Bonus Elit, didapatkan jika kamu mendapatkan 5 orang di bulan tersebut (berlaku kelipatan, sebesar 500.000)
Bonus Level, didapatkan jika ada satu garis di tiap downline
Bonus Posisi, merupakan bentuk incentive jika omset tim/downlinemu mencapai titik tertentu (omset tim merupakan total nilai investasi di tim kamu)
Bonus Royalti, merupakan bonus yang diberikan jika kamu memiliki omset 350 juta dan diatas Gold Agency

```python
import pandas as pd
import math

data_status = pd.DataFrame()
data_status['Bonus Rekrut'] = ["Done",0.25,1]
data_status['Bonus Elit'] = ["Done",0.1,1.37]
data_status['Bonus Level'] = ["In Progress",0.4,1.37]
data_status["Bonus Posisi"] = ["Done",0.82,1.37]
data_status["Bonus Royalti"] = ["Done",0.05,1.37]
data_status_output = data_status.transpose()
data_status_output.columns = ['Analysis',"Bonus","NPV"]
data_status_output["NPV Conversion"] = [1] + [0.32 for i in range(4)] 
data_status_output["Value"] = data_status_output["Bonus"]*data_status_output["NPV"]*data_status_output["NPV Conversion"]
print(data_status_output["Value"].sum())
data_status_output.round(2)
```

1. Bonus Rekrut
Bonus yang diberikan jika anda merekrut orang secara langsung, bonus yang diberikan sebesar 25%. Yang dibagikan kurang lebih

4/7 nya ditahun pertama (1/7 dari bagian downline)
2/7 nya ditahun kedua (1/14 dari bagian downline)
1/7 nya ditahun ketiga (1/28 dari bagian downline)

2. Bonus Elit
Bonus yang diberikan jika anda merekrut banyak orang secara langsung dalam satu bulan, bonus yang diberikan adalah 10% NPV. Yang dibagikan jika

5-9 mitra = 1 bagian
(lebih/sama) 10 mitra = 2 bagian
Formulanya = Bagian Anda*(10%NPV)/Total Bagian Elit di 3iCAR

3. Bonus Posisi
Bonus yang diberikan jika jumlah omzet setahun pertama memenuhi kriteria level tersebut, ini memiliki 40% NPV (jumlah dari semua bonus alokasi ditabel/N = 40%)

Formulanya = Persentase(N)*NPV/Mitra Kualifikasi Bonus Posisi
