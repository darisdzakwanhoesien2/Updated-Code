# The task is to analysis : 
1. Generate an Erdos-Renyi (random) network with 100 nodes, and plot the degree distribution;

```python
#import NetworkX and pyplot
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
from math import factorial
#import seaborn as sns
```

```python
p = 0.05  #5% chance of creating an edge
n = 100
G = nx.erdos_renyi_graph(n,p)
plt.suptitle("Erdos-Renyi Network",fontsize=16,y=1.0)
plt.title("100 Nodes with 5% chance of creating link.",fontsize=12)
#plt.suptitle("100 Nodes with 50% chance of creating link.")
nx.draw(G,with_labels=True)
plt.show()
```

```python
deg = [G.degree(n) for n in G.nodes()]
deg.sort()
tp = np.random.poisson(n*p, 100)
bins= [-0.5,0.5,1.5,2.5,3.5,4.5,5.5,6.5,7.5,8.5,9.5,10.5,11.5,12.5]
plt.style.use('ggplot')
plt.xlabel("No. of degrees(links)")
plt.ylabel("Count")
plt.hist(tp,bins,density=True)
plt.hist(deg,bins,density=True, alpha = 0.5,color = 'b')
plt.show()
```
