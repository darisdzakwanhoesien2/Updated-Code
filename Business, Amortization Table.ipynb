```python
def table_amortization_2(initial,invest,ROI,paid,interest_rate):
    import pandas as pd
    data_intro = pd.DataFrame()
    
    initial = initial
    paid = paid
    interest = initial*interest_rate/(12*100)
    payment = paid - interest
    remain = initial - payment
    
    invest = invest
    profit = invest*ROI/(12*100)
    total_invest = invest + profit
    
    valuation = remain - invest
    
    interest2 = remain*interest_rate/(12*100)
    payment2 = paid - interest2
    remain2 = remain - payment2
    total_invest2 = invest*2 + profit
    
    valuation2 = remain - total_invest2
    
    initial_list = [0,initial,paid,interest,payment,remain,invest,0,invest,valuation]
    second_list = [1,remain,paid,interest2,payment2,remain2,invest*2,profit,total_invest2,valuation2]
      
    data_intro[0] = initial_list
    data_intro[1] = second_list
    data_intro_new = data_intro.transpose()
    data_intro_new.columns = ['No',"Initial","Paid","Interest","Payment","Remain",'Invest','Profit','Total Invest','Valuation']
    return data_intro_new.round(2)
```

```python
def add_table_2(initiate_database,month,paid,invest,ROI,interest_rate):
    for i in range(month):
        initial = initiate_database['Remain'][i+1] #initial
        interest = initial*interest_rate/(12*100)
        payment = paid - interest
        remain = initial-payment
        
        profit = invest*(i+2)*ROI/(12*100) #ROI
        total_invest = invest*(i+3) + profit
        valuation = remain - total_invest
        
        initiate_database.loc[i+2] = [i+2,initial,paid,interest,payment,remain, invest*(i+3),profit,total_invest,valuation] #invest*2,profit,total_invest2,valuation2
    return initiate_database.round(2)
```

```python
def data_output(initial, allocation,paid,invest,ROI, interest_rate,month):
    initiate_database = table_amortization_2(initial,invest,ROI,paid,interest_rate)
    data = add_table_2(initiate_database,month,paid,invest,ROI,interest_rate)
    return data
def end_value(initial, allocation, paid, invest, ROI, interest_rate, month):
    data = data_output(initial, allocation, paid, invest, ROI, interest_rate, month)
    return data['Valuation'][month+1]
```

```python
paid = 50

initial, allocation = 50000, 2000
invest = allocation-paid

ROI, interest_rate = 6, 5
month = 30

print(end_value(initial, allocation, paid, invest, ROI, interest_rate, month))
data_output(initial, allocation,paid,invest,ROI, interest_rate,month)
```

```python
def graph(data):
    import matplotlib.pyplot as plt
    import numpy as np

    zero = [i for i in range(month+2)]
    plt.plot(data['No'], data['Remain'], label='Remaining Loan') #f : numpy
    plt.plot(data['No'], data['Valuation'], label='Valuation') 
    plt.plot(data['No'], data['Total Invest'], label='Total Invest') #g : numpy
    plt.plot(data['No'], zero, label='Zero Line')

    # f = data['Remain'].to_numpy()
    # g = data['Total Invest'].to_numpy()
    
    test = np.sign(data['Remain'] - data['Total Invest'])
    
    idx = np.argwhere(np.diff(test)).flatten()
    plt.plot(data['No'][idx], data['Remain'][idx], 'ro')
    
    plt.legend()
    print(idx)
    return plt.show()

graph(data)
```

```python
def saving(a,t,n):
  """This is a saving plan that you need to keep it up, a indicate your expected cash flow in the future, t is inflation rate prediction, n is the number of years that you need to survive"""
  r = 1 + t/100
  result = (a*r**(n+1)-a)/(r-1)
  return "With the inflation rate prediction of " + str(t) + " percent , expected current liabilities of " + str(a/12) + ". You are expected to have this capital without income " + str(result) + " and you are expected to save " + str(result/(n*12))

case1 = saving(60000000,10,50)
print(case1)
```
