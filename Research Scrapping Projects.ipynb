Two function created
1. link_to_txt(link)         : converting link into text (str)
2. clean_text(text,element)  : parsing journal data text (e.g. references)

```python
def link_to_txt(link):
    with urllib.request.urlopen(link) as response:
        html = response.read()
    soup = BeautifulSoup(html, "lxml")
    text_only = soup.get_text()
    return text_only

text_g_only = link_to_txt('https://en.wikipedia.org/wiki/CMOS')
text_g_only2 = link_to_txt('https://link.springer.com/article/10.1007/s42341-018-00092-5')
text_g_only3 = link_to_txt('https://doi.org/10.1007/s42341-018-0008-x')
```

```python
def references_specific_cleansing(data,split_1,split_2):
    names = ["References List",'Raw Link']
    names2 = ['Link','Junk 1','Junk 2']
    
    split_data2 = data['References List'].str.split(split_1)
    data2 = split_data2.to_list() 
    new_df2 = pd.DataFrame(data2, columns=names)
    
    split_data3 = new_df2['Raw Link'].str.split(split_2)
    data3 = split_data3.to_list()
    new_df3 = pd.DataFrame(data3, columns=names2)
    
    new_df4 = new_df3.drop(['Junk 1','Junk 2'], axis=1)
    # merging final output
    new_df5 = new_df2.join(new_df4)
    new_df6 = new_df5.drop(['Raw Link'],axis=1)
    return new_df6
    
new_g_df6 = references_specific_cleansing(data_info3,'https://','\n') #name set to be specific local
```
