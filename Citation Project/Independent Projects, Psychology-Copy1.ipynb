```python
#The parameter that is construct is the (a,b) = (pre-req, course) and the location of the graph 
#The automation are needed in two steps

# 1. Creating the web scrapping from the website (either become dataframe) and construct it into graph
# 2. Creating the graph creation 
```

```python
import networkx as nx
psychology_graph = nx.DiGraph()
pos = {}

def graph_creation_general(graph_initiate, pos, course_name, location_1, pre_reg, location_2):
    graph_initiate.add_edge(pre_reg,course_name)
    pos.update({course_name:location_1})
    pos.update({pre_reg:location_2})
    return graph_initiate
```

```python
course_list_hp2500, hp2500, pos_hp2500 = ['HP3501','HP4051','HP4091'], 'HP2500', (10,0) # HP2500-HP2600 : "HP4233","HP4263"
for i in range(len(course_list_hp2500)):
    psychology_graph = graph_creation_general(psychology_graph,pos,course_list_hp2500[i],(15,10*i),hp2500,pos_hp2500)

course_list_hp2600, hp2600, pos_hp2600 = ['HP3003','HP3302','HP3601','HP3702','HP4061','HP4232','HP4271',"HP4272","HP4273"], 'HP2600', (10,-10)
for i in range(len(course_list_hp2600)):
    psychology_graph = graph_creation_general(psychology_graph,pos,course_list_hp2600[i],(15,-10*(i+1)),hp2600,pos_hp2600)

# HP2600-HP2700 : HP3702, HP4272, HP4273
course_list_hp2700, hp2700, pos_hp2700 = ['HP3704','HP4102','HP4104','HP4105','HP4106'], 'HP2700', (10,-20)
for i in range(len(course_list_hp2700)):
    psychology_graph = graph_creation_general(psychology_graph,pos,course_list_hp2700[i],(15,-10*(i+10)),hp2700,pos_hp2700)    
    
nx.draw(psychology_graph, with_labels=True, font_weight='bold', pos=pos)
```
