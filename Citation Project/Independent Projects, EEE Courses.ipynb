```python
#The parameter that is construct is the (a,b) = (pre-req, course) and the location of the graph 
#The automation are needed in two steps

# 1. Creating the web scrapping from the website (either become dataframe) and construct it into graph
# 2. Creating the graph creation 
```

```python
import networkx as nx
eee_graph = nx.DiGraph()
pos = {}

def graph_creation_general(graph_initiate, pos, course_name, location_1, pre_reg, location_2):
    graph_initiate.add_edge(pre_reg,course_name)
    pos.update({course_name:location_1})
    pos.update({pre_reg:location_2})
    return graph_initiate
```

```python
course_list = ["EE3001","EE3010","EE3011","EE3012","EE3013","EE3019"]
pre_reg_course = ["EE2007","EE2001","EE2006","EE2010","EE2003","EE2002"]
for i in range(len(course_list)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list[i],(0,10*i),pre_reg_course[i],(-10,10*i))

eee_graph = graph_creation_general(eee_graph,pos,'EE3014',(0,-10),'EE2010',(-10, 30))
eee_graph = graph_creation_general(eee_graph,pos,'EE3015',(0,-20),'EE2001',(-10, -10))
eee_graph = graph_creation_general(eee_graph,pos,'EE4190',(0,-30),'EE2010',(-10, 30))

course_list_ee3012 = ['EE4152','EE4153','EE4188']
for i in range(len(course_list_ee3012)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee3012[i],(10,10*i),'EE3012',(0,30))

course_list_ee3019 = ['EE4303','EE4304','EE4341','EE4343','EE4344']
equal_value_1 = 'EE3019'
for i in range(len(course_list_ee3019)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee3019[i],(20,10*i),equal_value_1,pos.get(equal_value_1))

course_list_ee3013 = ['EE4613','EE4614','EE4645','EE4646','EE4694']
equal_value_2 = 'EE3013'
for i in range(len(course_list_ee3013)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee3013[i],(10,10*(i-5)),equal_value_2,pos.get(equal_value_2))

course_list_ee2010 = ['EE3012','EE3014','EE4190']
equal_value_3 = 'EE2010'
for i in range(len(course_list_ee2010)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee2010[i],(0,10*(i+7)),equal_value_3,pos.get(equal_value_3))

course_list_ee3010 = ['EE4503','EE4504','EE4530','EE4532','EE4534']
equal_value_4_1 = 'EE3010'
equal_value_4_2 = 'EE3015'
for i in range(len(course_list_ee3010)):
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee3010[i],(10,10*(i+3)),equal_value_4_1,pos.get(equal_value_4_1))
    eee_graph = graph_creation_general(eee_graph,pos,course_list_ee3010[i],(10,10*(i+3)),equal_value_4_2,pos.get(equal_value_4_2))


nx.draw(eee_graph, with_labels=True, font_weight='bold', pos=pos)
```
